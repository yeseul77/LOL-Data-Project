```python
item_url = 'https://ddragon.leagueoflegends.com/cdn/13.24.1/data/ko_KR/item.json'
```

```python
item_list = requests.get(item_url).json()
```

```python
item_df = pd.DataFrame(item_list['data']).transpose() # DataFrame을 생성하고 행과 열을 바꿈
```

```python
item_df.reset_index(inplace=True)  # 기존 인덱스를 열로 변환
```

```python
item_tmp = item_df[['index', 'name', 'into', 'gold', 'from', 'depth', 'inStore', 'requiredAlly', 'requiredChampion']]
```

```python
item_tmp['gold'] = [{ 'total': d['total'] > 1500 } for d in item_tmp['gold']] #구매골드가 1500이상이면 True
```

```python
item_tmp = item_tmp[item_tmp['gold'] != {'total': False}] #구매골드가 1500이상만 출력
```

```python
item_tmp_end = item_tmp[['index', 'name']]
```
